```python
import networkx as nx
from matplotlib import pyplot as plt
import datetime
```

```python
###############QUESTION 4 START#############################
q4 = nx.DiGraph()
q4.add_node('v1')
q4.add_node('v2')
q4.add_node('v3')
q4.add_node('v4')
q4.add_node('v5')
q4.add_node('v6')
q4Edges = [('v1', 'v2'),('v2', 'v1'), ('v3', 'v2'), ('v4', 'v3'), ('v3', 'v6'), ('v6', 'v3'), ('v5', 'v4'), ('v6', 'v5'), ('v5', 'v6')]
q4.add_edges_from(q4Edges)
undirected_non_reciprocal = q4.to_undirected(reciprocal=False)
nx.draw(undirected_non_reciprocal, with_labels=True, font_weight='bold')
plt.title("Undirected with Reciprocal=False")

```

```python
q4 = nx.DiGraph()
q4.add_node('v1')
q4.add_node('v2')
q4.add_node('v3')
q4.add_node('v4')
q4.add_node('v5')
q4.add_node('v6')
q4Edges = [('v1', 'v2'),('v2', 'v1'), ('v3', 'v2'), ('v4', 'v3'), ('v3', 'v6'), ('v6', 'v3'), ('v5', 'v4'), ('v6', 'v5'), ('v5', 'v6')]
q4.add_edges_from(q4Edges)
undirected_non_reciprocal = q4.to_undirected(reciprocal=True)
nx.draw(undirected_non_reciprocal, with_labels=True, font_weight='bold')
plt.title("Undirected with Reciprocal=True")
```

```python
q5 = nx.MultiDiGraph()
q5.add_nodes_from([('v1',{'Location': 'Salt Lake City', 'Job': 'Student', 'Age': 24}), ('v2',{'Location': 'Logan', 'Job': 'Student'}), ('v3',{'Gender': 'Female', 'Job': 'CEO'}), 
                   ('v4',{'Country': 'Mexico', 'Interests': 'Soccer'}), ('v5',{'Location': 'Phoenix', 'Job': 'Programmer', 'Interests': 'Coding'}), 
                   ('v6',{'Location': 'Las Vegas', 'Job': 'Musician', 'Interests': 'Music, Movies'})])
colors = {'v1': 'red', 'v2': 'red', 'v3': 'yellow', 'v4': 'green', 'v5': 'green', 'v6': 'yellow'}
nx.set_node_attributes(q5, colors, 'color')
color_map = [q5.nodes[n]['color'] for n in q5]
q5.add_weighted_edges_from([('v1','v2',8),('v2','v1',6),('v3','v2',8),('v3','v6',5),('v6','v3',2),('v4','v3',10),('v5','v4',7),('v6','v5', 13),('v5', 'v6',13)])
edge_labels = {(u, v): d['weight'] for u, v, d in q5.edges(data=True)}
node_labels = {}
pos = nx.spring_layout(q5) 
plt.figure(figsize=(5,5))
nx.draw(q5, pos, node_color=color_map, with_labels=True, node_size=400, connectionstyle="arc3, rad = 0.1")
nx.draw_networkx_edge_labels(q5, pos, edge_labels=edge_labels, font_color='black')

plt.show()
```
