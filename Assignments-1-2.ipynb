```python
import networkx as nx
from matplotlib import pyplot as plt
import datetime
```

```python
#########################QUESTION 2 START#########################
q2 = nx.Graph()
q2.add_node('v1')
q2.add_node('v2')
q2.add_node('v3')
q2.add_node('v4')
q2.add_node('v5')
q2.add_node('v6')
q2Edges = [('v1', 'v2'), ('v1', 'v6'), ('v2', 'v3'), ('v2', 'v4'), ('v3', 'v6'), ('v6', 'v5'), ('v5', 'v4'), ('v4', 'v3')]
q2.add_edges_from(q2Edges)
q2.add_weighted_edges_from([('v1','v2',5), ('v1','v6',4),('v3','v2',4),('v3','v6',3),('v4','v2',3),('v4','v3',1),('v5','v4',2),('v6','v5', 1)])
edge_labels = {(u, v): d.get('weight') for u, v, d in q2.edges(data=True)}
pos = {
    'v1': (1.8, 2),
    'v2': (1, 1.8),
    'v3': (1.2, 1),
    'v4': (1, 0),
    'v5': (1.8, 0.2),
    'v6': (2, 1.2)
}
shortest_paths = nx.shortest_path(q2, weight='weight', source='v1')
for destination, path in shortest_paths.items():
    if(destination != 'v1'):
        print(f"Shortest path from v1 to {destination}: {path}, Length = {nx.shortest_path_length(q2, weight='weight', source='v1', target=destination)}")
nx.draw(q2, pos, with_labels=True, font_weight='bold')
nx.draw_networkx_edge_labels(q2, pos, edge_labels=edge_labels, font_color='black')
plt.show()

#########################QUESTION 2 END##############################
```

```python
###############QUESTION 4 START#############################
q4 = nx.DiGraph()
q4.add_node('v1')
q4.add_node('v2')
q4.add_node('v3')
q4.add_node('v4')
q4.add_node('v5')
q4.add_node('v6')
q4Edges = [('v1', 'v2'),('v2', 'v1'), ('v3', 'v2'), ('v4', 'v3'), ('v3', 'v6'), ('v6', 'v3'), ('v5', 'v4'), ('v6', 'v5'), ('v5', 'v6')]
pos = {
    'v1': (1.8, 2),
    'v2': (1, 1.8),
    'v3': (1.2, 1),
    'v4': (1, 0),
    'v5': (1.8, 0.2),
    'v6': (2, 1.2)
}
q4.add_edges_from(q4Edges)
undirected_non_reciprocal = q4.to_undirected(reciprocal=False)
nx.draw(undirected_non_reciprocal, pos, with_labels=True, font_weight='bold')
plt.title("Undirected with Reciprocal=False")

```
